```python
import matplotlib.pyplot as plt
import matplotlib

import numpy as np
```

## Matplotlib
- Figure : 그림을 그릴 수 있는 영역 자체
- Subplot : 실제 그림이 그려지는 영역
    
### Matplotlib을 그리는 순서
1. Figure 설정
2. Subplot 설정

```python
# 새로운 종이 생성
fig = plt.figure( )

# 종이를 나누는 것
# (111) : 행번호, 열번호, 그림번호
ax = fig.add_subplot(111)

# 데이터
data = [0,1]

# 나눈 종이에 데이터 입력
ax.plot(data)

# 출력
plt.show( )
```

---

```python
# 새로운 종이 생성
fig = plt.figure()

# 새종이를 여러등분으로 나누는 방법
x1 = fig.add_subplot(221) # 2행2열의 첫번째 그래프
x2 = fig.add_subplot(222) # 2행2열의 두번째 그래프
x3 = fig.add_subplot(223) # 2행2열의 세번째 그래프

plt.show()
```

### 꺽은선 그래프 그리기

### 산포도 그래프 그리기
- 회귀분석 등 특정패턴을 찾을 때 유용한 그래프

```python
# 데이터 생성
np.random.seed(2)
x = np.arange(1,101)
y = 2 * x * np.random.rand(100)
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(x,y)

plt.show()
```

## 그래프 레이블 생성하기
- ax.bar(..,.. tick_label = 생성한 라벨이름)

```python
# 그래프와 라벨명 생성
labels = ['english', 'math']
x = [3,4]
y = [1,2]
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.bar(x,y,tick_label = labels)

plt.show()
```

### 가로로 누운 그래프 그리기
- ax.barh()

```python
# 데이터와 라벨명 생성
labels = ['english','math']

x = [3,4]
y = [1,2]
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.barh(x,y,tick_label=labels)

plt.show()
```

### 여러 데이터를 묶어 그리기

```python
# 데이터 생성
x = [1,2]
y1, y2, y3 = [1,2], [2,5], [3,7]
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

# 막대 두께
w = 0.2
ax.bar(x,y1,width=w, label= 'english')
ax.bar(np.array(x) + w, y2,width=w, label = 'math')
ax.bar(np.array(x) + w * 2,y3,width=w, label = 'korean')

ax.legend() # 그래프 색상에 관한 값을 표현
plt.show()
```

### 히스토그램 그리기

```python
# 데이터 생성
data1 = 100
data2 = 10
np.random.seed(0)
x = np.random.normal(data1,data2,1000)
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.hist(x)

plt.show()
```

### pie 차트 그리기

```python
plt.style.use('ggplot')
```

```python
# 라벨명과 사이즈 지정
labels = ['english', 'math', 'korean']
sizes = [45,30,25]
```

```python
fig = plt.figure(figsize=(3,3))

ax = fig.add_subplot(111)

ax.pie(sizes, labels=labels)

plt.show()
```
