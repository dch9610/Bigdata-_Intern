# 스쿨존 내 어린이 교통사고 건수

```python
import pandas as pd


import matplotlib.pyplot as plt
import matplotlib

import numpy as np

from matplotlib import pyplot as plt
```

```python
plt.rcParams['font.family'] = 'NanumGothic' # 윈도우(Windwos) 사용자
#plt.rcParams['font.family'] = 'AppleGothic' # 맥(Mac) 사용자
%matplotlib inline
```

```python
inschoolzone = pd.read_excel('C:/Users/user/Desktop/taas/도로종류별 스쿨존내 어린이(12세이하) 교통사고(1+2).xls',header=0)
```

```python
inschoolzone
```

```python
a = inschoolzone[:2]
a=a[['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']]
a
```

```python
count = list(a.iloc[1].values)[:]
count
```

# 막대그래프 시각화

```python
yearss = pd.read_csv('C:/Users/user/Desktop/taas/년도별 사고발생현황 전처리.csv')
yearss
```

```python
count1 = list(yearss.iloc[1].values)[1:]

count1
```

```python
count2 = list(yearss.iloc[0].values)[1:]
count2
```

```python
year = list(yearss.columns[1:])
year
```

```python
fig=plt.figure(figsize=(20,10))

ax=fig.add_subplot(111)

y = year
x1= count1
x2= count2

plt.bar(y,x2)
ax.bar(y,x1,bottom=x2)
ax.set_xlabel('년도')
ax.set_ylabel('발생건수')

plt.title('년도별 어린이 교통사고발생건수')
ax.legend(["스쿨존 외", "스쿨존 내 "], loc =0)

plt.show()
```
